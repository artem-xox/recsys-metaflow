```python
run_id = '16'
task_id = '106'
```

```python
ratings = Task(f'MovielensALSFlow/{run_id}/train/{task_id}', attempt=0)['ratings'].data
```

```python
model = Task(f'MovielensALSFlow/{run_id}/train/{task_id}', attempt=0)['model'].data
```

---

```python
movies_mapping = ratings[['itemID', 'title', 'genre']].drop_duplicates().set_index('itemID')
```

```python
def _similar_wrapper(item_id, N=10, model=model):
    return movies_mapping.loc[
        list(model.similar_items(item_id, N=N)[0])
    ]
```

```python
_similar_wrapper(1, N=20)
```
